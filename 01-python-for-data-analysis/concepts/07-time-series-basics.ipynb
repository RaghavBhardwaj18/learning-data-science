```python
# creating datetime and timestamp object data types
import numpy as np
import pandas as pd
from datetime import datetime

date1 = "12-05-2026"
print("date as string:", date1)
print("date as datetime object:", datetime.strptime(date1, "%d-%m-%Y"))
print("datetime object as string in different format:", datetime.strptime("2026-04-30 13:45:30", "%Y-%m-%d %H:%M:%S").strftime("%H:%M:%S %m-%d-%Y"))
```

```python
# shifting (leading and lagging data)
ts = pd.Series(np.random.standard_normal(5), index=pd.date_range("2026-05-01", periods=5, freq="D"))
print(ts)
```

```python
print(ts.shift(2))  # shifts time series data values with 2 positions ahead
```

```python
print(ts.shift(-3))  # shifts time series data values with 3 positions behind
```

```python
print(ts.shift(2, freq="D"))  # shifts time series data values along with frequency index update with 2 values ahead
```

```python
print(ts.shift(-1, freq="D"))  # shifts time series data values along with frequency index update with 1 value behind
```

```python
# time zone handling
# time zone localization and conversion

dates = pd.date_range("2026-01-01", periods=12, freq="M")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.index.tz)  # by default time zone of a time series is naive
```

```python
ts = ts.tz_localize("UTC")
print(ts)
```

```python
print(ts.tz_convert("America/New_York"))
```

```python
print(ts.tz_convert("Asia/Shanghai"))
```

```python
dates = pd.period_range("2025","2026", freq="A-DEC")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.asfreq("M", how="start"))
```

```python
print(ts.asfreq("D", how="end"))
```

```python
ts = ts.asfreq("A-JUN")
print(ts)
```

```python
dates = pd.period_range("2021","2026", freq="A-DEC")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.asfreq("A-JUN"))
```

```python
p = pd.Period("2021", freq="A-DEC")
print(p.asfreq("M"))
print(p.asfreq("M", how="start"))
print(p.asfreq("M", how="end"), "\n")

p = pd.Period("2021", freq="A-JUN")
print(p.asfreq("M"))
print(p.asfreq("M", how="start"))
print(p.asfreq("M", how="end"))
```

```python
# resampling time series
dates = pd.date_range("2026-01-01", periods=12, freq="MS")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("Q-MAR").mean())
```

```python
# downsampling (aggregating)

dates = pd.date_range("2024","2026", freq="M")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("A-DEC").sum())
```

```python
print(ts.resample("A-DEC").count())
```

```python
print(ts.resample("A-DEC").mean())
```

```python
# open high low close (ohlc) resampling
dates = pd.date_range("2026-01-01", periods=12, freq="T")
ts = pd.Series(np.arange(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("5min").sum())  # by default keeps closed as left
```

```python
print(ts.resample("5min", closed="right").sum())
```

```python
print(ts.resample("5min").ohlc())
```

```python
print(ts.resample("5min", closed="right").ohlc())
```

```python
# upsampling time series
import numpy as np
import pandas as pd
from datetime import datetime

dates = pd.date_range("2026-02", "2026-03", freq="MS")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("D").asfreq())
```

```python
print(ts.resample("D").asfreq().ffill())
```

```python
print(ts.resample("D").asfreq().ffill(limit=4))
```

```python
print(ts.resample("D").asfreq().bfill(limit=2))
```
