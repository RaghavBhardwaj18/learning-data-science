```python
# creating datetime and timestamp object data types
import numpy as np
import pandas as pd
from datetime import datetime

date1 = "12-05-2026"
print("date as string:", date1)
print("date as datetime object:", datetime.strptime(date1, "%d-%m-%Y"))
print("datetime object as string in different format:", datetime.strptime("2026-04-30 13:45:30", "%Y-%m-%d %H:%M:%S").strftime("%H:%M:%S %m-%d-%Y"))
```

```python
# defining a series with timeindex and values with date_range
dates = pd.date_range(start="2026-04-25", periods=6)
ts = pd.Series(np.random.standard_normal(6), index=dates)
print(ts)
print(type(ts))
print(type(ts.index))

print(ts.index[2])
print(ts[ts.index[2]])
print(ts["2026-04-26"])

# ts.index = ts.index.strftime("%Y-%m-%d")
# print(ts)
# print(ts.index[2])
# print(type(ts.index))

ts.index = ts.index.date
# ts.index = ts.index.normalize()
print(ts)
print(ts.index[2])
print(type(ts.index))
```

```python
# shifting (leading and lagging data)
ts = pd.Series(np.random.standard_normal(5), index=pd.date_range("2026-05-01", periods=5, freq="D"))
print(ts)
```

```python
print(ts.shift(2))  # shifts time series data values with 2 positions ahead
```

```python
print(ts.shift(-3))  # shifts time series data values with 3 positions behind
```

```python
print(ts.shift(2, freq="D"))  # shifts time series data values along with frequency index update with 2 values ahead
```

```python
print(ts.shift(-1, freq="D"))  # shifts time series data values along with frequency index update with 1 value behind
```

```python
# resampling time series
dates = pd.date_range("2026-01-01", periods=12, freq="MS")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("Q-MAR").mean())
```

```python
# downsampling (aggregating)

dates = pd.date_range("2024","2026", freq="M")
ts = pd.Series(np.random.standard_normal(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("A-DEC").sum())
```

```python
print(ts.resample("A-DEC").count())
```

```python
print(ts.resample("A-DEC").mean())
```

```python
# open high low close (ohlc) resampling
dates = pd.date_range("2026-01-01", periods=12, freq="T")
ts = pd.Series(np.arange(len(dates)), index=dates)
print(ts)
```

```python
print(ts.resample("5min").sum())  # by default keeps closed as left
```

```python
print(ts.resample("5min", closed="right").sum())
```

```python
print(ts.resample("5min").ohlc())
```

```python
print(ts.resample("5min", closed="right").ohlc())
```

```python
# Q2. Extract components
# year
# month
# day
# weekday
# df["month"] = df["date"].dt.month

df = pd.DataFrame(np.random.standard_normal((5,3)), index=["2026-04-05", "2026-05-05", "2026-06-05", "2026-07-05", "2026-08-05"], columns=["A","B","C"])
print(df)
print(df.index.dtype, "\n")

df.reset_index(inplace=True)
print("Updated dataset:")
print(df)
```

```python
print(df["index"].dtype)
df["index"] = pd.to_datetime(df["index"])
print(df["index"].dtype)
```

```python
df["year"] = df["index"].dt.year
df["month"] = df["index"].dt.month
df["day"] = df["index"].dt.day
df["weekday"] = df["index"].dt.weekday
print(df)
```

```python
# Q3. Create features
# weekend flag
# quarter
# day name

df["is_weekend"] = df["weekday"].isin([6,7]).astype(int)
df["quarter"] = df["index"].dt.quarter
df["day_name"] = df["index"].dt.day_name()
print(df)
```

```python
# Q4. Set index
# df.set_index("date", inplace=True)

df.set_index("index", inplace=True)
print(df)
```

```python
# Q5. Slicing
# df["2023"]
# df["2023-05"]

print(df.index.dtype)
print(df.loc["2026-05"])
```

```python
# Q6. Handle duplicate timestamps

# New timestamp as a dictionary
date = pd.to_datetime("2026-05-05")
values = list(np.random.standard_normal((3))) + [date.year, date.month, date.day, date.weekday(), int(date.weekday==6 or date.weekday==7), date.quarter, date.day_name()]
new_row = {key:value for key, value in zip(df.columns, values)}

new_index = [date]

new_row_df = pd.DataFrame([new_row], index=new_index)

# Concatenate without ignoring the index
df = pd.concat([df, new_row_df], ignore_index=False)

print(df)
```

```python
print(df.loc["2026-05"])
```

```python
df.reset_index(inplace=True)
```

```python
# Group by all categorical/date columns and average the numeric ones
df_merged = df.groupby(['index','year', 'month', 'day', 'weekday', 'is_weekend', 'quarter', 'day_name']).mean().reset_index()
print(df_merged)
```

```python
# Q16. Rolling mean
# df["rolling_mean"] = df["value"].rolling(7).mean()

df["rolling_mean_A"] = df["A"].rolling(7).mean().bfill(limit=7)
print(df)
```

```python
# Q17. Rolling std

df["rolling_std_C"] = df["C"].rolling(5).std().bfill(limit=5)
print(df)
```

```python
# Q18. Expanding window

df["expanding_window_A"] = df["A"].expanding(7).mean().bfill(limit=7)
print(df)
```
