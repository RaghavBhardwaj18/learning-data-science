**Numpy Basics**

**Creating Numpy arrays**

```python
# numpy arrays creation methods :
# array, asarray, arange, zeros, zeros_like, ones, ones_like, empty, empty_like, full, full_like, eye, identity
import numpy as np

# array : creates a numpy array simply by accepting a sequence of data
list1 = [1,2,3,4.5,5.2]
dict1 = {"Raghav":101,"Angad":102,"Jayant":103}
dict2 = {101:100,201:400,300:501}
tup1 = ((1,2),(2,1),(3,2),(3,1))
arr1 = np.array(list1)
arr2 = np.array(list(dict1.items()))
arr3 = np.array(tup1)
arr4 = np.array(list(dict2))

print(arr1, "dim:",arr1.ndim, "shape:",arr1.shape, "dtype:",arr1.dtype)
print(arr2, "dim:",arr2.ndim, "shape:",arr2.shape, "dtype:",arr2.dtype)
print(arr3, "dim:",arr3.ndim, "shape:",arr3.shape, "dtype:",arr3.dtype)
print(arr4, "dim:",arr4.ndim, "shape:",arr4.shape, "dtype:",arr4.dtype)
```

```python
# asarray --> doesn't creates a copy as the array method does
arr5 = np.asarray(list1)
print(arr5)

# ones, ones_like
list2 = [[1,2,3,4,5],[2,3,4,5,6]]
arr6 = np.ones((5,2))
arr7 = np.ones_like(list2)
print(arr6)
print(arr7)

# zeros, zeros_like
arr8 = np.zeros((2,3))
arr9 = np.zeros_like(list(dict2))
print(arr8)
print(arr9)

# arange
arr10 = np.arange(15)
print(arr10)

# full
full_arr = np.full((2,3),10)
print(full_arr)

# eye --> can create diagonal matrix as well and not just square matrix
eye_arr = np.eye(2,3)
print(eye_arr)

# identity --> can just create square matrix
identity_arr = np.identity(3)
print(identity_arr)
```

```python
# 🎯 Objective
# Simulate future stock prices using NumPy vectorization.

# 🧩 Dataset (Real-world)
# Use any from: Yahoo Finance, NSE India, Kaggle stock dataset
# Recommended: Reliance, TCS, NIFTY50

# 🛠 Tasks
# Phase 1 — Data Prep
# Load historical prices
# compute daily returns
# compute mean & std

# Phase 2 — Simulation
# Using NumPy:
# simulate 1000 future price paths
# horizon: 252 days
# use geometric Brownian motion

# Phase 3 — Analysis
# Compute:
# expected price
# worst-case scenario
# best-case scenario
# volatility bands

# Phase 4 — Visualization
# Plot:
# sample paths
# distribution histogram
# confidence interval bands

import numpy as np
import pandas as pd

data = pd.read_csv("data.csv")
print(data.head(),"\n",type(data),"\n")

data_array = np.array(data)
print(data_array[:8],"\n")

# print("daily differences:")
# print(data_array[:,[0,4]])
daily_diff = np.round(np.diff(data_array[:,-1].astype(float),axis=0),3)
# print(daily_diff[:5],"\n")
# print("daily returns:")
daily_returns = np.round(daily_diff/(data_array[:data_array.shape[0]-1,-1].astype(float)),3)
# print(daily_returns[:5],"\n")
daily_returns = daily_returns.reshape(-1,1)

print("date-wise daily returns:")
date_wise_daily_returns = np.concatenate((data_array[1:,0].reshape(-1,1),daily_returns),axis=1)
print(date_wise_daily_returns[:5],"\n")

# print("date-wise daily returns (in %):")
date_wise_daily_returns_pct = date_wise_daily_returns.copy()
date_wise_daily_returns_pct[:,-1] = np.round((date_wise_daily_returns_pct[:,-1]*100).astype(float),3)
# print(date_wise_daily_returns_pct[:5],"\n")

# arr1 = np.array([1,2,3,4,5])
# print(arr1)
# arr1 = arr1.reshape(-1,1)
# print(arr1)

# print("mean of daily returns:",np.round(np.mean(daily_returns,axis=0)*100,3))
# print("standard deviation of daily returns:",np.round(np.std(daily_returns,axis=0)*100,3))
print("mean of daily returns:",np.round(np.mean(date_wise_daily_returns_pct[:,-1],axis=0),3))
print("standard deviation of daily returns:",np.round(np.std(date_wise_daily_returns_pct[:,-1],axis=0),3))

print("Simulating 1000 future paths using Geometric Brownian Motion:")
# Data Preparation and Parameter Estimation
# Starting Price (S0)
# Annual drift (u) = average log - daily return
# Annual volatility (sigma) = std deviation of log - daily returns

ratio_daily_prices = data_array[1:,-1]/data_array[:data_array.shape[0]-1,-1]
print(np.round((ratio_daily_prices[:5].astype(float)),3))

log_ratio_daily_prices = np.log(ratio_daily_prices.astype(float))
print(np.round((log_ratio_daily_prices[:5]).astype(float),3),"\n")

drift = np.mean(log_ratio_daily_prices)
volatility = np.std(log_ratio_daily_prices)
So = data_array[-1,-1]
print("drift (mean of daily log returns):",np.mean(log_ratio_daily_prices))
print("volatility (std deviation of daily log returns):",np.std(log_ratio_daily_prices))
print("starting price (So, the most recent closing price):",data_array[-1,-1])

# The Simulation Setup
T = 252
t = 1
paths = 1000
Z = np.random.standard_normal((252,1000))

# GBM discrete formula : S(t+1) = S(t) x exp((drift - volatility**2/2)t + volatility*sqrt(t)Z)

# daily "multiplier" for every single step in every path
daily_growth = np.exp((drift - (volatility**2 / 2)) + volatility * Z)

future_price_paths = np.zeros((T+1,paths),dtype="float")
future_price_paths[0] = So
future_price_paths[1:] = daily_growth
# future_price_paths[1:] = np.cumprod(future_price_paths[0:]*np.exp((drift - (volatility**2/2))*t + volatility*np.sqrt(t)*np.random.choice(Z.flatten())))
future_price_paths = np.cumprod(future_price_paths,axis=0)
# print(np.round(future_price_paths[:5],3))
print(np.round(future_price_paths[-1:-4:-1],3))
print(future_price_paths.shape)

print("expected price after 1 year:",np.mean(future_price_paths[-1:],axis=1))
print("worst-case scenario:",np.min(future_price_paths[-1:],axis=1))
print("best-case scenario:",np.max(future_price_paths[-1:],axis=1))

import matplotlib.pyplot as plt

# plt.plot(future_price_paths[-1:,:4].T)
plt.plot(future_price_paths[:, :4])
# plt.hist(future_price_paths)
plt.hist(future_price_paths.flatten())


import matplotlib.pyplot as plt

# 5. Visualization
plt.figure(figsize=(12, 6))

# Plot all 1000 paths with low transparency (alpha) to see the "density"
plt.plot(future_price_paths, color='blue', alpha=0.05, linewidth=1)

# Plot the Mean Path (the average of all simulations) in Red
mean_path = np.mean(future_price_paths, axis=1)
plt.plot(mean_path, color='red', linewidth=2, label='Expected Mean Path')

# Formatting the Chart
plt.title(f"Nifty 50: 1,000 Future Price Paths Simulation (GBM)\nStart Price: {So:.2f}", fontsize=14)
plt.xlabel("Days into Future", fontsize=12)
plt.ylabel("Index Value", fontsize=12)
plt.axhline(So, color='black', linestyle='--', alpha=0.5, label='Current Price')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Quick summary statistics
final_prices = future_price_paths[-1]
print(f"Mean Expected Price after 1 year: {np.mean(final_prices):.2f}")
print(f"95% Confidence Interval: [{np.percentile(final_prices, 2.5):.2f}, {np.percentile(final_prices, 97.5):.2f}]")

# # 1. Calculate the 2.5th and 97.5th percentiles for each day (axis=1)
# lower_bound = np.percentile(future_price_paths, 2.5, axis=1)
# upper_bound = np.percentile(future_price_paths, 97.5, axis=1)
# median_path = np.percentile(future_price_paths, 50, axis=1) # Median (50th percentile)

# # 2. Plot the main paths (keep alpha low to avoid clutter)
# plt.plot(future_price_paths, color='gray', alpha=0.03, linewidth=1)

# # 3. Plot the Median and Confidence Interval Bounds
# plt.plot(median_path, color='blue', linewidth=2, label='Median Path (50th Percentile)')
# plt.plot(lower_bound, color='red', linestyle='--', linewidth=1.5, label='95% Confidence Interval (Boundaries)')
# plt.plot(upper_bound, color='red', linestyle='--', linewidth=1.5)

# # 4. Fill the area between the bounds to create a "shadow" or "fan"
# plt.fill_between(range(T + 1), lower_bound, upper_bound, color='red', alpha=0.1, label='95% Probability Zone')

# # 5. Add text labels for the final expected range
# plt.text(T, upper_bound[-1], f' Bull Case: {upper_bound[-1]:.0f}', color='red', verticalalignment='bottom')
# plt.text(T, lower_bound[-1], f' Bear Case: {lower_bound[-1]:.0f}', color='red', verticalalignment='top')

# plt.title("Nifty 50: 95% Probability Cone (GBM Simulation)", fontsize=14)
# plt.legend(loc='upper left')
# plt.show()

```
