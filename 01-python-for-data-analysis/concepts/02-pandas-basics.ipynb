**Function Application and Mapping**

```python
# define a function and apply it to a dataframe through the apply method (both axes)
# use numpy ufuncs on dataframes
# use dataframes in built methods as well
# use applymap function as well

import pandas as pd
import numpy as np

# creating a dataframe of students marks with rows as students and columns as subjects
df1 = pd.DataFrame({"John":np.random.randint(1,101,5),"Aray":np.random.randint(1,101,5),"Sam":np.random.randint(1,101,5),"Jay":np.random.randint(1,101,5)})
df1 = df1.T
df1.index.name = "Students"
df1.columns.name = "Subjects"
df1.columns = ["English","Hindi","Maths","Science","Social Science"]
print(df1)

# numpy ufuncs
# print(np.sqrt(df1))
# print(np.exp(df1))

# defining a function manually to return max and min marks in subjects (subjects-wise)
def f1(x):
  return pd.Series([x.max(),x.min()], index=["max","min"])

# defining a function manually to return deviation from median marks (considering median to be 50), students-wise
def f2(x):
  return abs(x-50)

def f3(x):
  return (x.max()-x.min())/100

print(df1.apply(f1))
print(df1.apply(f2, axis="columns"))
print(df1.apply(f3, axis="columns"))

series1 = pd.Series([1,2,3,4,5])
print(series1.map(np.exp))

def f4(x):
  return x+10

# print(df1.applymap(f4))
print(df1.map(f4))
```
