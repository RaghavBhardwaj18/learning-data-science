**Handling Missing Data**

```python
# missing values representation in pandas
import numpy as np
import pandas as pd

series1 = pd.Series([0.2,5.6,np.nan,4,None])
print(series1)

print(series1.isna())
print(series1.isna().sum())
print(series1.isna().value_counts())
```

Filtering missing data

```python
# we use dropna() method to filter out missing data from series or dataframe
# key parameters of dropna() method are: how (how="any", how="all") , thresh (computes minimum number of non-NA values along the indicated axis to give the filtered result) , inplace (True/False)

series1 = pd.Series([3.4,2.5,6.7,np.nan,8.6,None])
print(series1)
print(series1.isna())

print(series1[series1.notna()])
print(series1[~series1.isna()])
series1.dropna(inplace=True)
series1
```

```python
df1 = pd.DataFrame(np.random.standard_normal((5,3)))
print(df1)

df1.iloc[2:4,:2] = np.nan
df1.iloc[1:3,2] = np.nan
df1.iloc[4,1] = np.nan
print(df1)

# filtering out missing data with just using pd.isna function and boolean indexing
print(df1[df1.iloc[:,0].notna()]) # removes rows where column 0 has null values
df1.dropna() # drops rows with any null values
df1.dropna(how="all") # drops rows with all null values
df1.dropna(thresh=2) # keeps rows (since default axis=0) with minimum 2 non-NA values
df1.dropna(thresh=3, axis=1) # keeps columns with minimum 3 non-NA values
```

Filling in missing data

```python
# we can fill missing data using fillna() method
# key parameters of fillna() method are : value to fill missing values with (constant or dictionary for column-wise specification) , method ("ffill" or "bfill") , axis ("index" or "columns" / 0 or 1) , limit (maximum number of values to fill with during forward fill or backward fill when doing interpolation using the method argument)
print(df1)

df1.fillna(0)
replace_missing_values = {0:5.2, 1:2.4, 2:3}
df1.fillna(replace_missing_values)
df1.fillna(method="ffill")
df1.fillna(method="bfill", limit=1, axis=1)

# filling missing values column wise with their mean values
df1.fillna(df1.mean(axis=0))
# filling missing values row wise with their mean values
# df1.fillna(df1.mean(axis=1), axis=1)  # --> results in error due to non-alignment of row index values with column labels of the dataframe
# df1.T.fillna(df1.mean(axis=1), axis=0).T
df1 = df1.apply(lambda row:row.fillna(row.mean()), axis=1)
df1
```

```python
# map method can be used for specifying a dictionary to map values of a column with or a function for transforming it's values
import numpy as np
import pandas as pd

df1 = pd.DataFrame({"English":np.random.randint(25,101,10), "Hindi":np.random.randint(25,101,10), "Maths":np.random.randint(25,101,10), "Social Science":np.random.randint(25,101,10)}, index=list("Rohan Akshay Sohan Shilpa Ishita Abhay Vikrant Aniket Dikshita Sandra".split()))
print("Student dataframe:")
print(df1)

# assigning a "english fluency" title to every student based on their english marks
english_marks_categories = [(df1["English"]<=30), (df1["English"]>=31) & (df1["English"]<=50), (df1["English"]>=51) & (df1["English"]<=70) , (df1["English"]>=71) & (df1["English"]<=90) , (df1["English"]>=91) & (df1["English"]<=100)]
english_titles = ["Beginner", "On way to fluency", "Fluent", "Viber", "Professional"]
english_marks_level_dict = np.select(english_marks_categories, english_titles, default="")
# print(english_marks_level_dict)
df1["english_level"] = english_marks_level_dict
# print(df1)

# assigning cities from where students belong using map function
stud_city_mapping = {key:value for key,value in zip(df1.index, np.random.choice(list("Delhi Uttar_Pradesh Bengaluru Mumbai Pune Odisha Chennai".split()), 10))}
df1["City"] = df1.index.map(stud_city_mapping)

# alternative way for passing a function to map method instead of a dictionary
def map_stud_city(stud):
  return stud_city_mapping[stud]

df1["City_from_function"] = df1.index.map(map_stud_city)
print(df1)
```

```python
# Q10.
# detect outliers using IQR

data = np.random.randint(0,101,10)
data[5:10:2] = [312,-5,500]
print(data)

# IQR method for outlier detection
# Outliers are data values that fall below the value of Q1-1.5*IQR or fall above the value of Q3+1.5*IQR
# IQR = Q3-Q1

q3 = np.quantile(data,0.75)
q1 = np.quantile(data,0.25)
# iqr = np.quantile(data,0.75) - np.quantile(data,0.25)
iqr = q3 - q1
# print("outlier values present in data:", data[(data<(-1.5*iqr)) | (data>1.5*iqr)])
print("outlier values present in data:", data[(data<(q1-(1.5*iqr))) | (data>(q3+(1.5*iqr)))])
```

```python
# Q11.

# remove outliers
# cap outliers

data_outliers = np.array(data[(data<(q1-(1.5*iqr))) | (data>(q3+(1.5*iqr)))])
data = np.array(data)
print("outliers in data:", data_outliers)

data_copy = data.copy()

print("data with outliers:", data)
data = data[~np.isin(data,data_outliers)]
print("data without outliers:", data)

print("original data copy:", data_copy)
data_copy = np.where(np.isin(data_copy, data_outliers), 1000, data_copy)
print("data copy after capping outlier values:", data_copy)
```
