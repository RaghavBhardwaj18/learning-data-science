```python
 # importing required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# creating a figure
fig = plt.figure()

# adding 3 subplots
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

# creating a line plot
ax1.plot(np.random.standard_normal(50).cumsum(), color="black", linestyle="dashed")

# creating a histogram
ax2.hist(np.random.standard_normal(1000), bins=20, alpha=0.3)

# creating a scatter plot
ax3.scatter(np.arange(50), np.arange(50) + 3*np.random.standard_normal(50))
print(fig)
```

```python
# creating the same subplots with axes iterator
fig, axes = plt.subplots(2,2)

axes[0,1].plot(np.arange(10))
axes[1,0].hist(np.random.standard_normal(100), bins=10, alpha=0.5)
print(fig)
```
