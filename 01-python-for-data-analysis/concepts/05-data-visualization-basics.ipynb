```python
# Q4.
# highlight max value
# add text annotations

fig = plt.figure()
ax = fig.add_subplot()

data = np.random.uniform(0,100,50).cumsum()
ax.plot(data, linestyle="dotted", color="purple", alpha=1)

max_val = np.max(data)
# print(max_val)

line = ax.lines[0]
# print(line)
x_data, y_data = line.get_xdata(), line.get_ydata()
# print(x_data)
# print(y_data)

target_y = max_val
x_target = np.interp(target_y, y_data, x_data)
print("max value coordinates: ","(","x:",x_target,",","y:",target_y,")")

# ax.text(x_target, target_y, "max-value data point")
ax.annotate("Max Value",
            xy = (x_target,target_y),
            xytext = (x_target-20, target_y),
            arrowprops = dict(facecolor="black", shrink=0.05),
            fontsize=8)

target_y2 = np.median(data)
x_target2 = np.interp(target_y2, y_data, x_data)
print("median value coordinates: ","(","x:",x_target2,",","y:",target_y2,")")
ax.annotate("Median value",
            xy=(x_target2,target_y2),
            xytext=(x_target2-5,target_y2+500),
            arrowprops=dict(facecolor="red", shrink=0.05),
            fontsize=9)
```

```python
# 🟢 Q5
# Plot:
# revenue over time
# Using Olist dataset

orders = pd.read_csv("olist_orders_dataset.csv")
orders_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
products_translated = pd.read_csv("product_category_name_translation.csv")

# print(orders.head())
# print(orders_items.head())
# print(products.head())
# print(products_translated.head())

# print(orders["order_status"].unique())
orders = orders.loc[orders["order_status"]=="delivered",["order_id","order_purchase_timestamp"]]
print(orders.head())

orders_items = orders_items.loc[:,["order_id","product_id","price","freight_value"]]
print(orders_items.head())

products = products.loc[:,["product_id","product_category_name"]]
print(products.head())

print(products_translated.head())
```

```python
# print(products.columns)
# print(products_translated.columns)
products_df = pd.merge(products, products_translated, on="product_category_name", how="inner")
products_df.drop("product_category_name", axis=1, inplace=True)
products_df.rename(columns={"product_category_name_english":"product_category"}, inplace=True)
print(products_df.head())
```

```python
orders_items_df = pd.merge(orders_items, products_df, on="product_id", how="inner")
print(orders_items_df.head())
```

```python
orders_df = pd.merge(orders, orders_items_df, on="order_id", how="inner")
orders_df.drop("product_id", axis=1, inplace=True)
print(orders_df.head())
```

```python
orders_df["total_value"] = orders_df["price"] + orders_df["freight_value"]
print(orders_df.head())
```

```python
# creating a line plot for revenue over time
import matplotlib.ticker as ticker

fig = plt.figure()
ax = fig.add_subplot()

orders_df_copy = orders_df.copy()
orders_df_copy = orders_df_copy[:100]

# orders_df_copy['order_purchase_timestamp'] = pd.to_datetime(orders_df_copy['order_purchase_timestamp'])
# orders_df_copy['order_purchase_timestamp'] = orders_df_copy['order_purchase_timestamp'].dt.date
# orders_df_copy['order_purchase_timestamp'] = orders_df_copy['order_purchase_timestamp'].str[:10]
# print(orders_df_copy["order_purchase_timestamp"].dtype)
# print(orders_df_copy.head())
ax.plot(orders_df_copy["order_purchase_timestamp"],orders_df_copy["total_value"])
ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
# ax.set_xticks(np.arange())
```

```python
# bar plot of revenue per category
fig = plt.figure()
ax = fig.add_subplot()

# revenue_per_category = orders_df.groupby("product_category")["total_value"].sum()
orders_per_category = orders_df.groupby("product_category")["order_id"].count()
# print(orders_per_category)

orders_per_category[:20].plot.bar(ax=ax)
ax.set_xlabel("product categories")
ax.set_ylabel("total orders placed")
```

```python
fig = plt.figure()
ax = fig.add_subplot()

revenue_per_category = orders_df.groupby("product_category")["total_value"].sum()
# print(revenue_per_category)

revenue_per_category[:20].plot.bar(ax=ax)
ax.set_xlabel("product categories")
ax.set_ylabel("total price value")
```

```python
# 🔴 Q7 (IMPORTANT)

# Sort values before plotting
fig = plt.figure()
ax = fig.add_subplot()
revenue_per_category.sort_values(ascending=False, inplace=True)
revenue_per_category[:20].plot.bar(ax=ax)
ax.set_xlabel("product categories")
ax.set_ylabel("total price value")
```
