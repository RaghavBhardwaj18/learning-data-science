```python
# Q1. Create MultiIndex

# create Series with multi-level index
# create DataFrame with hierarchical rows

import numpy as np
import pandas as pd

series1 = pd.Series([1,2,3,4,5], index=[["one","one","one","two","two"],[1,2,3,1,2]])
print(series1)

df1 = pd.DataFrame(np.arange(24).reshape((6,4)), index=pd.MultiIndex.from_arrays([[1,1,1,2,3,3],["one","two","three","four","four","five"]]), columns="first second third fourth".split())
print(df1)
```

```python
# Q6. Basic Merge

# Given:

# customers table
# orders table

# Perform:
# inner join
# left join
# right join
# outer join

# Sample Customers Table
customers_data = {
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Chicago'],
    'age': [25, 30, 35, 40, 28]
}
customers = pd.DataFrame(customers_data)

# Sample Orders Table
orders_data = {
    'order_id': [101, 102, 103, 104, 105, 106],
    'customer_id': [1, 2, 1, 3, 2, 1], # 1 links to Alice, 2 to Bob, 3 to Charlie
    'order_date': ['2023-01-01', '2023-01-05', '2023-01-10', '2023-02-01', '2023-02-05', '2023-03-01'],
    'amount': [150.50, 200.00, 50.00, 300.25, 120.00, 75.00]
}
orders = pd.DataFrame(orders_data)

print("Customers Table:\n", customers)
print("\nOrders Table:\n", orders)
```

```python
# inner join
print(pd.merge(customers, orders, on="customer_id"))
```

```python
# left join
print(pd.merge(orders, customers, on="customer_id", how="left"), "\n")
print(pd.merge(customers, orders, on="customer_id", how="left"))
```

```python
# right join
print(pd.merge(orders, customers, on="customer_id", how="right"), "\n")
print(pd.merge(customers, orders, on="customer_id", how="right"))
```

```python
# outer join
print(pd.merge(orders, customers, on="customer_id", how="outer"), "\n")
print(pd.merge(customers, orders, on="customer_id", how="outer"))
```

```python
# Q7.

# merge on multiple columns
# merge with different column names

import pandas as pd

# Customer data: Includes unique IDs and shared location info
customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 104],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'region': ['North', 'South', 'North', 'West'],
    'last_active': ['2023-01', '2023-05', '2023-03', '2023-06']
})

# Order data: Multiple orders per customer, overlapping 'region' column
orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4, 5],
    'cust_id': [101, 101, 102, 103, 105], # 105 doesn't exist in customers
    'region': ['North', 'North', 'East', 'North', 'West'],
    'amount': [250, 150, 300, 450, 100]
})

print("customers data:\n", customers)
print("orders data:\n", orders)
```

```python
print(pd.merge(customers, orders, left_on=["customer_id","region"], right_on=["cust_id","region"]))
```

```python
print(pd.merge(customers, orders, left_on="customer_id", right_on="cust_id"))
```

```python
# Q12.

# use combine_first()

# Example:
# fill missing values using another dataset

# Table A: Primary Customer Records
customers = pd.DataFrame({
    'cust_id': [101, 102, 103, 104],
    'region': ['North', 'South', 'East', 'West'],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'loyalty_points': [500, np.nan, 300, 150]
})

# Table B: Order History (Multiple entries for some customers, different regions)
orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4, 5],
    'cust_id': [101, 102, 101, 103, 105], # 105 is a new customer
    'region': ['North', 'South', 'North', 'East', 'North'],
    'amount': [250, 150, 450, 300, 100]
})

# Table C: Supplemental Data for combine_first
# (Contains missing data for existing IDs or new info)
supplemental = pd.DataFrame({
    'loyalty_points': [500, 200, 300, 150], # Fills Bob's missing points
    'name': ['Alice', 'Bob', 'Charlie', 'David']
}, index=[101, 102, 103, 104])

print("customers data:\n", customers)
print("orders data:\n", orders)
print("supplemental:\n", supplemental)
```

```python
customers.set_index("cust_id").combine_first(supplemental)
```

```python
# Q17.

# Dataset: sales

# Create:

# revenue per region
# revenue per category
# region × category matrix

# Sample sales data
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-01', '2023-02-01', '2023-02-02', '2023-02-01'],
    'Region': ['East', 'West', 'East', 'North', 'West', 'North'],
    'Product': ['Apples', 'Oranges', 'Bananas', 'Apples', 'Oranges', 'Bananas'],
    'Sales': [100, 200, 150, 110, 120, 160],
    'Quantity': [10, 20, 15, 12, 13, 16]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
print(df)
```

```python
df["Revenue"] = df["Sales"]*df["Quantity"]
```

```python
print(df)
```

```python
# revenue per region
print(df.pivot_table(index="Region", values="Revenue", aggfunc="sum"))
```

```python
# revenue per category
print(df.pivot_table(index="Product", values="Revenue", aggfunc="sum"))
```

```python
# region x category matrix
print(df.pivot_table(index="Region", columns="Product"))
```

```python
print(df.pivot_table(index="Region", columns="Product", values=["Sales", "Revenue"]))
```

```python
print(df.pivot_table(index="Region", columns="Product", values="Sales"))
```
